```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
from datetime import date

today = date.today()

today = today.strftime("%Y/%m/%d")
```

```python
df = df[df['Date']<today]
```

```python
df = df[['Wk','Date','Home','Away','HomeGoals','AwayGoals','HXG','AXG']]
```

```python
df = df.reset_index()
```

```python
df = df.drop(columns = 'index')
```

```python
data = df[['Home','Away','HXG','AXG','Wk']]
```

```python
data
```

```python
home_df = data.copy()
home_df = home_df.melt(id_vars = ["Wk", "Home", "Away"])
home_df["venue"] = "H"
home_df.rename(columns = {"Home":"team", "Away":"opponent"}, inplace = True)
home_df.replace({"variable":{"HXG":"xG_for", "AXG":"xG_ag"}}, inplace = True)
```

```python
home_df
```

```python
away_df = data.copy()
away_df = away_df.melt(id_vars = ["Wk", "Away", "Home"])
away_df["venue"] = "A"
away_df.rename(columns = {"Away":"team", "Home":"opponent"}, inplace = True)
away_df.replace({"variable":{"AXG":"xG_for", "HXG":"xG_ag"}}, inplace = True)
```

```python
df = pd.concat([home_df, away_df]).reset_index(drop = True)
```

```python
teams = data['Home'].unique()
```

```python
teams
```

```python
teams = data['Home'].unique()

color1 = ['#001688','#012b55','#111471','#000000','#00793f','#fcd011','#e21e26','#1434A4','#303392','#fcd20f','#0a4f98','#000000','#0f4d84','#54863C']
color2 = ['#cd0102','#3262a8','#fac10c','#ad9059','#000000','#000000','#ffffff','#2e2e83','#C2C0DD','#000000','#e4c521','#77712B','#f0d013','#A4AE9C']


mydict = {}

for index,value in enumerate(teams):
    mydict[value] = [color1[index],color2[index]]
```

```python
mydict
```

```python
Tot = 14
Cols = 4

Rows = Tot // Cols 

if Tot % Cols != 0:
    Rows += 1
    
Position = range(1,Tot + 1)
```

```python
from matplotlib import font_manager

font_dirs = [r'C:\FontsCHAR']
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)

for font_file in font_files:
    font_manager.fontManager.addfont(font_file)

# set font
plt.rcParams['font.family'] = 'ABeeZee'
```

```python
fig = plt.figure(figsize=(10,8), dpi = 200,facecolor="#f9f1f1")
for index,value in enumerate(teams):
    df = pd.concat([home_df, away_df]).reset_index(drop = True)

    df = df[df["team"] ==  value].reset_index(drop = True)
    df = df.sort_values(by = "Wk").reset_index(drop = True)

    # xG conceded and xG created
    Y_for = df[df["variable"] == "xG_for"].reset_index(drop = True).drop(columns = 'Wk')
    Y_ag = df[df["variable"] == "xG_ag"].reset_index(drop = True).drop(columns = 'Wk')
    X = pd.Series(range(len(Y_for)))

    # Compute the rolling average (min_periods is used for the partial average)
    # Here we're using a 10 game rolling average
    Y_for = Y_for.rolling(window = 5, min_periods = 0).mean()
    Y_ag = Y_ag.rolling(window = 5, min_periods = 0).mean()
    
    
    
    ax = fig.add_subplot(Rows,Cols,Position[index])
    ax.set_facecolor('#f9f1f1')
    ax.plot(X,Y_for)      # Or whatever you want in the subplot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines["left"].set_color("grey")
    ax.spines["bottom"].set_color("grey")
    ax.grid(
    visible = False, 
    lw = 0.75,
    ls = ":",
    color = "lightgrey")
    line_1 = ax.plot(X, Y_for, zorder = 4,color = mydict[value][0])
    line_2 = ax.plot(X, Y_ag,'-.', zorder = 4, color = mydict[value][1])
    ax.set_ylim(bottom = 0, top = 2)
    ax.plot(
    [26,26],
    ls = ":",
    lw = 1.25,
    color = "grey",
    zorder = 2)
    
    ax.fill_between(
    X, 
    Y_ag["value"],
    Y_for["value"], 
    where = Y_for["value"] >= Y_ag["value"], 
    interpolate = True,
    alpha = 0.55,
    zorder = 3,
    color = mydict[value][0])
    
    ax.fill_between(
    X, 
    Y_ag["value"],
    Y_for["value"], 
    where = Y_ag["value"] > Y_for["value"], 
    interpolate = True,
    alpha = 0.55,
    zorder = 3,
    color = mydict[value][1])
    
    ax.tick_params(
    color = "grey", 
    length = 5, 
    which = "major", 
    labelsize = 6,
    labelcolor = "grey",
    zorder = 3)
    
    ax.set_xticks(np.arange(0, len(X)+1, 5))
    ax.xaxis.set_major_locator(ticker.MultipleLocator(13))
# Set y-axis major tick positions to only 0.5 xG multiples.
    ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))

    # Title and subtitle for the legend
    ax.set_title(label = value,loc = 'left',size = 'large',ha = 'center',in_layout = True,color = '#181D31')

plt.subplots_adjust(left=0.3, right=0.9, bottom=0.3, top=0.9)   

fig_text(
    x = 0, y = 1.07,
    s = "Expected goals <created> and <conceded> in Greek Superleague 2022-2023",
    color = "#181D31",
    size = 12
)

fig_text(
    x = 0, y = 1.05,
    s = "5 match rolling average | Straight line for xG-For, Dotted line for xG-Against",
    color = "#181D31",
    size = 9
)

fig_text(
    x = 0, y = 1.025,
    s = "Visual by kharzzz | @KidChapo22",
    color = "#181D31",
    size = 6
)


plt.tight_layout()
plt.savefig('GSL_20022023.png', dpi = 600, bbox_inches='tight')
plt.show()
```
